# Handling Null Values

```python
import pandas as pd
import numpy as np

coffee = pd.read_csv("./data_latihan/coffee.csv")
coffee.head()
```

```python
coffee["Price"] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 4.99)
```

```python
coffee.head()
```

```python
coffee.loc[[2,3], 'Units Sold'] = np.nan
coffee
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate().round()
coffee
```

```python
coffee.loc[[2,3], 'Units Sold'] = np.nan
coffee
```

```python
coffee = coffee.dropna(subset=['Units Sold'])
coffee
```

```python
coffee.loc[[6,8], 'Units Sold'] = np.nan
coffee
```

```python
coffee[coffee['Units Sold'].isna()]
```

```python
coffee[coffee['Units Sold'].notna()]
```
