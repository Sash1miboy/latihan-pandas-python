# Filtering Data

```python
import pandas as pd
bios = pd.read_csv("./data/bios.csv")
results = pd.read_parquet("./data/results.parquet")
```

```python
bios.loc[bios["height_cm"] > 215, ["name", "height_cm"]]
```

```python
bios[(bios["height_cm"] > 220) & (pd.to_datetime(bios["born_date"]) <= pd.to_datetime("2025-01-01"))]
```

```python
bios[bios["name"].str.contains("Michael")]
```

```python
bios[bios["born_country"].isin(["USA", "GER", "FRA"]) & bios["name"].str.startswith("Michael")] 
```

## Query Functions
