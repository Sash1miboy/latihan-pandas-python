# Accessing Data with Pandas

```python
import pandas as pd
```

```python
coffee = pd.read_csv("./data_latihan/coffee.csv")
results = pd.read_parquet("./data/results.parquet")
bios = pd.read_csv("./data/bios.csv")
```

```python
# sample akan mengambil data secara random
coffee.sample(10)
```

## How to Access a Single Data

```python
# loc[[rows], [columns]] untuk mencari single data
# dengan menggunakan rows index dan columns menggunakan nama headers 

coffee.loc[[0,1,3], ["Day", "Units Sold"]]
```

```python
# iloc[[rows], [columns]] untuk mencari single data
# dengan menggunakan index rows dan index columns
results.iloc[[123]]
```

```python
# Untuk akses columns
results[["year", "as"]]
```

```python
coffee["Day"] == "Monday"
```

```python
coffee.loc[:, ["Coffee Type"]]
```

```python
coffee.iloc[0, 0]
```

## Manipulate data with iloc & loc

```python
coffee.loc[[1], ["Units Sold"]] = 10
coffee.head()
```
