# Adding/Removing Columns

```python
import pandas as pd
coffee = pd.read_csv("./data_latihan/coffee.csv")
coffee
```

```python
# Add Columns Paling simple tetapi membuat semua row memiliki nilai yang sama
coffee["Price"] = 4.9
coffee
```

```python
import numpy as np

coffee["new_price"] = np.where(coffee["Coffee Type"] == "Espresso", 3.99, 5.99)
```

```python
# coffee.drop(columns=["price"], inplace=True)

# Pake diantara dua ini

# coffee = coffee.drop(columns=["Price"])

# tipe tipe drop
# drop_duplicates -> untuk menghapus duplicates value di columns dengan menggunakan subset=[] untuk memilih kolum yang ingin di drop
# dropna -> untuk drop missing value dalam data

coffee
```

```python
coffee = coffee.drop(columns=['Price'])
```

```python
coffee["Revenue"] = coffee["Units Sold"] * coffee ["new_price"]
coffee
```

```python
coffee = coffee.rename(columns={'new_price':'Price'})
```

```python
bios = pd.read_csv("./data/bios.csv")
bios
```

```python
bios_new = bios.copy()
bios_new
```

```python
bios_new['first_name'] = bios_new['name'].str.split(" ").str[0]
bios_new
```

```python
bios_new.query('first_name == "Michael" and born_city == "Chicago"')
```

```python
bios_new["born_datetime"] = pd.to_datetime(bios_new["born_date"])
bios_new["died_datetime"] = pd.to_datetime(bios_new["died_date"])
bios_new.head()
```

```python
bios_new["born_year"] = bios_new["born_datetime"].dt.year
bios_new["died_year"] = bios_new["died_datetime"].dt.year
bios_new[["name", "born_year", "died_year"]].head()
```

```python
bios_new.to_csv("./data/bios_born_died.csv", index=False)
```
