# Advanced Functionality

```python
import pandas as pd
import numpy as np
import pyarrow as pa 
```

```python
coffee = pd.read_csv("./data_latihan/coffee.csv")
bios = pd.read_csv("./data/bios.csv")
```

```python
coffee["Price"] = np.where(coffee["Coffee Type"] == "Espresso", 3.99, 5.99)
coffee["Revenue"] = coffee["Price"] * coffee["Units Sold"]
coffee.head()
```

```python
bios["height_rank"] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values('height_rank').head()
```

```python
coffee['cumulative_revenue'] = coffee['Revenue'].cumsum()
```

```python
coffee = coffee.rename(columns={'cumulative_revenue': 'Cumulative Revenue'})
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
